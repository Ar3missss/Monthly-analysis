```python
import pandas as pd
import openpyxl
import sqlite3
```

```python
data = pd.read_excel('data/Book1.xlsx' , sheet_name = 'MAR-2026' , skiprows=2)
## data = data.dropna(how='all')
## data = data.drop(index=1048572)
data = data[data['Date'].notna()]
data = data[['Date','Details','Income','Expenses','Balance']]
data[['Income','Expenses']]= data[['Income','Expenses']].fillna(0)
data = data.reset_index(drop=True)

conn =sqlite3.connect('personal_finance.db')
data.to_sql('Monthly_Expenses',conn,if_exists= 'replace',index=False) # in if_exists u can also use replace append and error according to situation
print('Done')
conn.close()
```
