```python
import pandas as pd
import openpyxl
import sqlite3
```

```python
data = pd.read_excel('data/Book1.xlsx' , sheet_name = 'MAR-2026' , skiprows=2)
## data = data.dropna(how='all')
## data = data.drop(index=1048572)
data['Details'] = data['Details'].str.lower().str.strip()
data = data[data['Date'].notna()]
data = data[['Date','Details','Income','Expenses','Balance']]
data[['Income','Expenses']]= data[['Income','Expenses']].fillna(0)
data = data.reset_index(drop=True)

conn =sqlite3.connect('personal_finance.db')
data.to_sql('Monthly_Expenses',conn,if_exists= 'replace',index=False) # in if_exists u can also use replace append and error according to situation
print('Done')
conn.close()

data.to_csv('clean_data.csv',index=False)
```

```python
feb = pd.read_excel('data/Book1.xlsx', sheet_name = 'FEB-2026', skiprows=3)
feb['Details'] = feb['Details'].str.lower().str.strip()
feb = feb[feb['Date'].notna()]
feb= feb[['Date','Details','Income','Expenses','Balance']]
feb[['Income','Expenses']]= feb[['Income','Expenses']].fillna(0)
feb['Income']= feb['Income'].astype('int64')

feb_conn = sqlite3.connect('personal_finance.db')
feb.to_sql('Monthly_Expenses',feb_conn,if_exists='append',index=False)
print('Done')
feb_conn.close()

feb.dtypes
feb
```
